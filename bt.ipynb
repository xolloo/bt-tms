```python
A = 27
n = 3

x = A
epsilon = 1e-8
iterations = 0

while True:
      x_next = (1 / n) * ((n - 1) * x + A / (x ** (n - 1)))
      if x == x_next or (x - x_next) <= epsilon:
            break
      x = x_next
      iterations += 1


print(x)
print(iterations)
```

```python
A = 40

for n in range(2, 27):
      x = round(A / (n * (n * 10 / 100)), 8)
      print(x, end=" | ")
      epsilon = 1e-8
      iterations = 0

      while True:
            x_next = (1 / n) * ((n - 1) * x + A / (x ** (n - 1)))
            if x == x_next or (x - x_next) <= epsilon:
                  if not iterations:
                        x += (x * 10 / 100)
                        continue
                  break
            x = x_next
            iterations += 1


      print(x, end=" -> ")
      print(iterations)
```
